## Bar Chart

```python
#plotting chart
sns.barplot(data=penguins, x='species', y="flipper_length_mm", hue='island')
plt.legend(loc="best")
plt.title("Flipper length for Species(by Island)")
plt.show()
```

**Bar Chart 설명**  

**Title : Flipper length for Species (by Island)**

이 차트는 펭귄 종에 따른 날개 길이를 나타내며, 섬 (Island)에 따라 분할되어 표시한다.

* **X-Axis (X축)**: Species (종) <br>
 X축은 펭귄의 종(species)을 나타낸다. <br>
Adelie, Chinstrap, Gentoo 세 가지 종이 있음을 알 수 있다.

* **Y-Axis (Y축)**: Flipper Length (날개 길이, 밀리미터) <br>
 Y축은 펭귄의 날개 길이를 밀리미터 단위로 나타낸다. <br>
 Y축 값은 각 종에 대한 평균 날개 길이를 나타낸다. <br>

* **Hue (색상)**: Island (섬) <br>
 각 종별로 다른 색상을 사용하여, 펭귄의 종에 따른 날개 길이를 섬(Island)별로 구분하고 있다.

* **Adelie 종** : 이 종은 Biscoe, Dream 및 Torgersen 섬 모두에서 발견된다.  <br>
Adelie 종의 평균 날개 길이는 섬(Island)별로 볼 때 Torgersen에서 가장 길었고, <br>
Dream, Biscoe 순으로 미미하게 짧아지는 형상을 볼 수 있다. <br>

* **Chinstrap 종** : Chinstrap 종은 Dream 섬에서만 발견되었다. <br>
이 종은 다른 섬에서는 관찰되지 않는다. 

* **Gentoo 종** : Gentoo 종은 Biscoe 섬에서만 발견되었고, 다른 섬에서 관찰되지 않았다. <br>

종별 날개의 길이를 비교해보았을 때 Adelie 종의 평균 날개 길이가 가장 짧은 것을 확인할 수 있다. <br>
그 다음으로 Chinstrap 종,  Gentoo 종 순으로 날개의 평균 길이가 길다는 사실을 확인할 수 있다.

## Scatter Chart

```python
sns.scatterplot( x = 'bill_length_mm',y='bill_depth_mm', data = penguins, hue = 'sex', 
                style = 'sex', size = 'body_mass_g', palette = 'Dark2')
plt.title("The relation Bill length and Bill depth (with Body mass)")
plt.show()
```

**Satter Cart 설명** 

**Title: The relation Bill length and Bill depth (with Body mass)** <br>

이 차트는 산점도를 통해 부리 길이, 부리 두께, 몸무게 및 성별 간의 관계를 시각화한 것이다. <br>

* **X-Axis (X축)**: Bill Length (부리 길이, 밀리미터) <br>
 X축은 펭귄의 부리 길이를 밀리미터 단위로 나타낸다. <br>

* **Y-Axis (Y축)**: Bill Depth (부리 두께, 밀리미터) <br>
 Y축은 펭귄의 부리 두께를 밀리미터 단위로 나타낸다. <br>

* **Hue (색상)**: Sex (성별) <br>
 색상을 통해 펭귄의 성별(sex)을 구분하고 있다. <br>
 이것은 산점도 내에서 성별에 따른 차이를 시각화한다.<br>

* **Style (스타일)**: Sex (성별) <br>
 스타일의 차이를 통해 성별 구분을 더 알아보기 쉽게 나타냈다.<br>

* **Size (크기)**: Body Mass (몸무게, 그램) <br>
 펭귄의 몸무게에 따라 점 크기에 차이를 두어, <br>
 크기의 변화를 통해 몸무게와 부리의 길이, 부리 두께 간의 관계를 표현하였다. <br>

**무게와 부리 길이의 관계** : <br>
산점도에서 무게가 무거운 펭귄일수록 부리의 길이가 긴 쪽에 더 많이 분포하는 것을 확인할 수 있다. <br>
 이는 몸무게와 부리 길이 간의 양의 관계를 시각화한 것이다. <br>

**무게와 부리 두께의 관계**: <br>
상대적으로 무게가 가벼운 펭귄들은 부리의 두께가 두꺼운 쪽에 분포하는 모습을 확인할 수 있으며, <br>
무게가 무거운 펭귄들은 부리가 얇은 쪽에 분포하는 것을 확인할 수 있다. <br>
이를 통해 전반적으로, 몸무게와 부리 두께 간의 관계가 음의 관계임을 알 수 있다.<br>

**성별에 따른 차이**: <br>
상대적으로 몸무게가 가벼운 무리와 무거운 무리로 나누어서 보았을때, <br>
전반적으로 암컷(female) 펭귄의 경우가 수컷(male) 펭귄에 비해 부리의 두께가 얇은 것을 확인할 수 있다.  <br>
이를 통해 성별 간 부리 특성의 차이를 확인할 수 있다. <br>

**무게가 가벼운 펭귄들과 무거운 펭귄들로 나누어 보았을 때 분포 현상** : <br>
상대적으로 무게가 가벼운 펭귄들의 분포가 더 다양하며, <br>
이는 몸무게가 가벼운 펭귄들 사이에 부리 길이와 부리 두께에 다양성이 있음을 나타낸다.<br>

무게가 무거운 펭귄들의 경우, 몸무게와 부리의 길이, 부리 두께 간의 관계가 상대적으로 몰려 있으며, <br>
부리 길이와 부리 두께의 범위가 좁게 집중되어 있음을 알 수 있다.

**Box plot 설명**

X 축은 펭귄의 종(species)을 나타낸다. <br>
Adelie, Chinstrap, Gentoo 세 가지 종이 있음을 알 수 있다.

Y축은 펭귄의 부리의 두께를 밀리미터 단위로 나타낸다.  <br>
y값은 box plot을 통해 펭귄 종별 부리 두께의 분포를 보여준다.

* **Hue (색상)**: Sex (성별) <br>
색상을 통해 펭귄의 성별(sex)을 구분하고 있다. <br>
box plot 내에서 성별에 따른 차이를 시각화한다. <br>

**펭귄 종 간 부리의 두께 차이:** <br>
Adelie, Chinstrap, Gentoo 종 모두에서 male(수컷) 펭귄의 부리가 female(암컷)보다 두껍다는 것을 확인할 수 있다. <br>
Adelie, Chinstrap, Gentoo 종 순서대로 박스의 형태가 짧아지는 것을 확인할 수 있다.<br>
이는 Adelie 종은 부리 두께가 가장 다양하며 범위가 넓고,<br>

Chinstrap 종은 중간 정도의 다양성, Gentoo 종은 부리 두께가 상대적으로 일정한 범위에 분포한다는 것을 의미한다. <br>
전체적인 box plot의 분포를 살펴보면, Adelie 종과 Chinstrap 종은 부리 두께가 비슷한 분포선상에 놓인다는 것을 알 수 있다. <br>
이는 두 종 사이의 부리 두께가 비슷하다는 것을 의미한다. <br>

gentoo 종의 경우 두 종보다 상대적으로 얇은 부리를 가지고 있음을 확인할 수 있다.  <br>

**Adelie 종의 부리 두께:**  <br>
Adelie 종의 경우, male과 female 모두에서 이상치(outlier) 값이 존재한다.<br>
이것은 부리의 두께가 일반적인 범위를 벗어난 데이터 포인트를 나타내며, 부리의 두께에 변별력을 준다.<br>
Adelie 종의 male 펭귄의 경우 중앙값보다 부리가 두꺼운 경우가 더 많다는 것을 확인할 수 있다.<br>

**Chinstrap 종의 부리 두께:** <br>
Chinstrap 종의 female 펭귄의 경우 중앙값보다 부리가 얇은 경우가 더 많다는 것을 확인할 수 있다. <br>


**Gentoo 종의 부리 두께:** <br>
Gentoo 종의 부리의 두께는 다른 두 종에 비해 상대적으로 분산이 작고, 박스가 짧게 그려진 것을 볼 수 있다. <br>
앞서 언급했듯이 이것은 부리의 두께가 Gentoo 종 내에서 상대적으로 일정한 범위 내에서 분포되고 있다는 것을 의미한다. <br>

## Joint Plot

```python
sns.jointplot(x='bill_length_mm', y='flipper_length_mm', data=penguins, hue = 'island')
plt.suptitle("The relationship between Bill length and Flipper length (by Island)")
plt.show()
```

**Join Plot 설명** <br>

**Title: The relationship between Bill length and Flipper length (by Island)** <br>

이 차트는 부리 길이와 날개 길이에 따른 펭귄의 분포를 나타낸다. (섬별 비교) <br>

* **X-Axis (X축)**: Bill Length (부리 길이, 밀리미터) <br>
 X축은 펭귄의 부리 길이를 밀리미터 단위로 나타낸다. <br>
부리 길이가 서식지에 따라 어떻게 변하는지 시각적으로 확인 가능하다. <br>

* **Y-Axis (Y축)**: Flipper Length (날개 길이, 밀리미터) <br>
Y축은 펭귄의 날개 길이를 밀리미터 단위로 나타낸다. <br>
날개 길이가 부리 길이와 어떤 관계를 가지는지 확인 가능하다. <br>

* **Hue (색상)**: Island (섬) <br>
 색상을 통해 펭귄의 서식지를 구분한다. <br>
 이를 통해 서로 다른 섬에 서식하는 펭귄 집단의 부리 길이와 날개 길이를 비교할 수 있다.<br>

* **부리 길이의 분포**: </br>
그래프에서는 각 섬별로 부리 길이의 분포를 확인할 수 있다.  <br>
Dream 섬에서 부리 길이가 가장 넓고 고르게 분포하고 있는 것을 알 수 있고, <br>
Biscoe 섬에서는 부리 길이가 긴 펭귄이 더 많이 서식하는 것으로 나타난다. <br>
Torgersen 섬에서는 부리 길이가 상대적으로 짧은 펭귄들이 많이 서식하는 것으로 보인다. <br>

* **날개 길이의 분포**: <br>
Biscoe 섬에는 날개 길이가 긴 펭귄이 많이 서식하는 것을 확인할 수 있으며, 날개의 길이가 가장 넓게 분포한다. <br>
Torgersen과 Dream 섬에서는 Biscoe 섬에 비해 날개 길이가 상대적으로 짧은 펭귄들이 서식하는 것을 알 수 있다. <br>

 이 그래프를 통해 서로 다른 섬에서 서식하는 펭귄 집단 간의 부리 길이, 날개 길이, <br>
 부리 길이와 날개 길이 사이의 관계를 비교하고 분석할 수 있다.<br>
 부리 길이와 날개 길이 사이의 상관 관계를 확인했을 때, 전체적으로 큰 상관 관계가 없다고 판단하였다.<br>
 그러나 Dream 섬을 제외하고 보았을 때, 부리 길이가 길수록 날개 길이가 긴 양상을 확인할 수 있다. <br>

하지만, 개별 데이터 포인트를 확인하는 것은 어려운데, strip plot은 개별 데이터 포인트를 표시하여 더 직관적인 데이터 분포를 파악할 수 있다. <br>
또한, 상대적으로 각 종마다 성별 간의 차이를 더 자세히 파악할 수 있다. 

## Stacked Histograms

```python
sns.histplot(data = penguins, x = 'body_mass_g', bins = 15, hue = 'species', multiple='stack')
plt.title('Distribution of body_mass')
plt.show()
```

**Stacked Histograms 설명** <br>

**Title: Distribution of body_mass** <br>

이 차트는 각 펭귄 종(species)별로 체중의 분포를 나타낸다. <br>

* **X-Axis (X축)**:  Body Mass (몸무게, 그램)

  X축은 펭귄의 몸무게를 그램 단위로 나타낸다. <br>
 몸무게 분포를 히스토그램으로 확인할 수 있다. <br>

* **Y-Axis (Y축)**: Count (수) <br>
 Y축은 펭귄의 수를 나타낸다. <br>

* **Hue (색상)**: Species (종) <br>
 색상을 통해 펭귄의 종(species)을 구분한다. <br>
 Adelie, Chinstrap, Gentoo 세 가지 종이 있음을 알 수 있다. <br>

* **Bins (구간)** : 데이터를 15개의 구간(bins)으로 나누어 분포를 시각화하였다. <br>

* **종별 몸무게 분포**:  <br>
Gentoo 종의 몸무게 분포는 다른 종들에 비해 오른쪽으로 더 치우쳐져 있으며, <br>
상대적으로 큰 몸무게를 가지는 개체가 많은 것을 알 수 있다. <br>

  Adelie와 Chinstrap 종은 비슷한 몸무게 분포를 가지는 것을 확인할 수 있다. <br>

  Adelie 종과 Gentoo 종은 각각의 종의 분포를 확인할 때, 몸무게의 분포가 이어지는 양상을 보인다. <br>
  하지만, Chinstrap 종은 왼쪽에서 첫 번째와 세 번째 막대 그래프 사이의 몸무게 범위에 해당하는 개체가 없음을 알 수 있다. <br>
  이는 Adelie 종에 outer 개체가 존재함을 나타낸다. <br>

* **Adelie 종의 수**: </br>
Adelie 종은 다른 두 종에 비해 훨씬 많은 수의 개체가 존재하는 것을 알 수 있다. <br>
Adelie 종의 몸무게 분포가 그래프에서 높은 빈도로 나타나는 것을 확인할 수 있다. <br>
